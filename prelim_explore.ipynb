```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col_names = ['id', 'click', 'hour', 'C1', 'banner_pos', 'site_id', 
             'site_domain', 'site_category', 'app_id', 'app_domain', 
             'app_category', 'device_id', 'device_ip', 'device_model',
             'device_type', 'device_conn_type', 'C14','C15','C16','C17','C18','C19','C20','C21']
```

```python
train = pd.read_csv('training.csv', names=col_names)
```

```python
train['hour'] = pd.to_datetime(train['hour'], format='%y%m%d%H')
```

```python
sns.countplot(data=train, x='click')
```

```python
cat_vars = list(train.select_dtypes(exclude=np.number).columns)
```

```python
cat_vars
```

```python
for col in train.columns:
    print('Column: ' + col)
    plt.figure(figsize=(8,4))
    plt.hist(train[col], edgecolor='black')
    plt.show()
```

```python
for col in cat_vars:
    print(col)
    print(train[col].value_counts())
    print(train[col].value_counts(normalize=True)*100)
    sns.countplot(x=col, data=train)
    plt.show()
```

```python
time_train = train.set_index('hour').sort_index()
```

```python
# plotting defaults
plt.rc('figure', figsize=(20, 7))
plt.style.use('seaborn-whitegrid')
plt.rc('font', size=16)

time_train.resample('H')[['click']].sum().plot()
```

```python
time_train['hour_of_day'] = time_train.index.strftime('%H')
```

```python
time_train.groupby('hour_of_day').click.mean().plot()
```

Most click throughs at around 1am and 3pm.

```python
time_train['day_of_week'] = time_train.index.strftime('%A')
```

```python
time_train.groupby('day_of_week').click.mean().plot()
```

```python
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday']
day_of_week_stats = time_train.groupby('day_of_week').click.mean().reindex(days).plot()
```

```python
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday', 'Sunday']
day_of_week_stats = time_train.groupby('day_of_week').click.sum().reindex(days)
```

```python
time_train.groupby('day_of_week').mean()
```
